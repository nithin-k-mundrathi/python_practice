Load a Phi-3 Model

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model Microsoft phi-3
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda", # where to store the Model
    torch_dtype="auto",
    trust_remote_code=True,
)
```

Load thee Phi-3 Tokenizer

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

Create a Pipeline to Generate text from Phi3 Model

```python
from transformers import pipeline

# Create a pipeline using the Model and Tokenizer
generator = pipeline(
    "text-generation", # Task at Hand
    model=model, # Model - phi3
    tokenizer=tokenizer, # Tokenizer of the Model
    return_full_text=False, # Not needed to Return Full Text
    max_new_tokens=500, # max tokens for the Model to Generate
    do_sample=False # Choose most Common token instead of using Sampling Strategy
)
```

Create a Prompt

```python
# The prompt (user input / query)
messages = [
    {
     "role": "user",
     "content": "Create a Product name for causal shirt with black stripes and white in colour for brand ralph lauren"}
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```

## Compare Tokenizers

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
```

```python
text = """
English and CAPITALIZATION &$% 532672
🎵 鸟
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*59=6000
"""
```

```python
# use sub-word Models - WordPiece
show_tokens(text, "bert-base-uncased")
```

```python
# use sub-word Models - WordPiece
show_tokens(text, "bert-base-cased")
```

```python
# Use sub-word Models -  BPE
show_tokens(text, "gpt2")
```

```python
# BPE -  Text to text Transform Transformers
show_tokens(text, "google/flan-t5-small")
```

# Train a Gensim Model for Song Recommendation

```python
import pandas as pd
from urllib import request

# Get the playlist dataset file
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# Parse the playlist dataset file. Skip the first two lines as
# they only contain metadata
lines = data.read().decode("utf-8").split('\n')[2:]

# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]

# Load song metadata
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])
songs_df = songs_df.set_index('id')
```

```python
print(playlists[0]),
print(playlists[10])
```

```python
from gensim.models import Word2Vec

# Train our Word2Vec model
model = Word2Vec(
    playlists, vector_size=10, window=20, negative=50, min_count=1, workers=8
)
```

```python
song_id = 2172
# Ask the model for songs similar to song #2172
model.wv.most_similar(positive=str(song_id))
```

```python
print(songs_df.iloc[5586])
```

```python
import numpy as np

def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id),topn=5)
    )[:,0]
    return  songs_df.iloc[similar_songs]

# Extract recommendations
print_recommendations(3317)
```
